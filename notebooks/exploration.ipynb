```python
import pandas as pd
import matplotlib.pyplot as plt #Sirve para visualizacion de datos

#Cargamos los datos limpios

data_path = "../data/processed/cleaned_products.csv" #Ruta archivo con datos a leer
df  = pd.read_csv(data_path) #Cargamos los datos del archivo csv en un DF

#Analisis y visualizacion

print("Basic Statistics")
print(df.describe()) #Mostramos un resumen estadisticos de los datos en el DF

#Histograma de precios

plt.hist(df["price"], bins=20, edgecolor="black") #Creamos un histograma de la columna PRICE con 20 barras y bordes negros
plt.title("Distribucion de precios") #Añadimos un titulo al histograma
plt.xlabel("Precio") #Añadimos una etiqueta el eje X
plt.ylabel("Frecuencia") #Añadimos una etiqueta el eje Y
plt.show() #Mostramos el histograma.

#Productos mas caros
print("Productos mas Caros: ") 
highestPrices = df.nlargest(5, "price")
print(highestPrices)
```
